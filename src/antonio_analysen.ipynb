```python
import pandas as pd
import matplotlib.pyplot as plt

# Beispiel-Daten
data = pd.DataFrame({
    'Jahr': [2019, 2020, 2021, 2022, 2023],
    'Verkäufe': [150, 200, 250, 300, 400]
})

# Plot
plt.figure(figsize=(8, 5))
plt.plot(data['Jahr'], data['Verkäufe'], marker='o')
plt.title('Jährliche Verkäufe')
plt.xlabel('Jahr')
plt.ylabel('Verkäufe')
plt.grid(True)
plt.show()
```
